```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
path = r'/home/vaibhavsharma/Desktop/DS23/hv/Time Series/data_engineer_sa_indicators'

csv_files = glob.glob(path + '/*.csv')
csv_files
```

```python
path_quarterly = csv_files[2]

df_quarterly = pd.read_csv(path_quarterly)
df_quarterly.drop(columns= ['Unnamed: 4'], inplace= True)

df_quarterly['PERIOD'] = pd.to_datetime(df_quarterly['PERIOD'])
df_quarterly.set_index(['PERIOD'], inplace= True)

df_quarterly.head()
```

```python
len(df_quarterly)
```

```python
# No. of unique indicators (quarterly)

len(df_quarterly['INDICATOR_NAME'].unique())
```

```python
df_quarterly = df_quarterly.pivot_table(index='PERIOD',columns='INDICATOR_NAME',values='ACTUAL_VALUE')
```

```python
df_quarterly.sort_index(ascending= True)
df_quarterly.tail()
```

```python
# Columns where NaN values are less than 10%

cols = [col for col in df_quarterly.columns if (((df_quarterly[col].isnull().sum() / len(df_quarterly)) * 100) < 70)]
cols, len(cols)
```
